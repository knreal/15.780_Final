```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

#Read in dataset
df = pd.read_csv('Clean_Dataset.csv')
df.head()
```

```python
#generate list of flight numbers, modify flight numbers to match URL
flights = df["flight"].tolist()
flightnos = []
for flight in flights:
    x = flight.replace('6E-', 'IGO').replace('UK-', 'VTI').replace('SG-', 'SEJ').replace('I5-', 'IAD').replace('G8-', 'GOW').replace('AI-', 'AIC')
    if x not in flightnos:
        flightnos.append(x)
```

```python
#Flight IGO6062 causes error, scrape the 1304 flights before it
URLbase = "https://flightaware.com/live/flight/"
craft = []
for flightno in flightnos[:1304]:
    URL = URLbase+flightno
    print(URL)
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, "html.parser")
    aircraft = str(soup.find(id ="mainBody").find("div", class_="flightPageAdUnit")).split("aircraft_type', '")[1].split("').setTargeting")[0]
    craft.append(aircraft)
print(len(craft))
#Should return 1304
```

```python
#Flight AIC9894 also causes error, scrape until 1446th flight
URLbase = "https://flightaware.com/live/flight/"
craft2 = []
for flightno in flightnos[1305:1446]:
    URL = URLbase+flightno
    print(URL)
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, "html.parser")
    aircraft = str(soup.find(id ="mainBody").find("div", class_="flightPageAdUnit")).split("aircraft_type', '")[1].split("').setTargeting")[0]
    craft2.append(aircraft)
print(len(craft2))
#Should return 141
```

```python
#Scrape remaining flights
URLbase = "https://flightaware.com/live/flight/"
craft3 = []
for flightno in flightnos[1447:]:
    URL = URLbase+flightno
    print(URL)
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, "html.parser")
    aircraft = str(soup.find(id ="mainBody").find("div", class_="flightPageAdUnit")).split("aircraft_type', '")[1].split("').setTargeting")[0]
    craft3.append(aircraft)
print(len(craft3))
#should return 114
```

```python
#Manual lookup shows that IGO6062 is an A320 and AIC9894 is not found
#merge everything into one list
aircraftlist = craft+['A320']+craft2+['']+craft3
```

```python
#Generate list of unique aircraft
unique = []
for thing in aircraftlist:
    if thing not in unique:
        unique.append(thing)
unique
```

```python
#Create counts of each 
counts = [0]*len(unique)
for j in aircraftlist:
    for i in range(len(unique)):
        if j == unique[i]:
            counts[i]+=1
    
    
```

```python
#Create dictionary
aircraftcounts = dict(zip(unique, counts))
aircraftcounts
```

```python
#Create dataframe
aircraftdf = pd.DataFrame(list(zip(flightnos, aircraftlist)),
               columns =['Flight', 'Aicraft'])
aircraftdf.head()
```
